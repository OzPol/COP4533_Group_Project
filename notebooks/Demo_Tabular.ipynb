```python
import random

# Non-Increasing Heights
def generate_non_increasing_cases():
    cases = []
    max_platform_width = 10
    for _ in range(10):
        n = random.randint(5, 10)
        heights = sorted([random.randint(1, 10) for _ in range(n)], reverse=True)
        widths = [random.randint(1, max_platform_width) for _ in range(n)]
        case = (n, max_platform_width, heights, widths, "Expected: TBD platforms, Total Height: TBD, Sculptures per Platform: TBD")
        cases.append(case)
    return cases


# Strictly Increasing Heights
def generate_increasing_cases():
    cases = []
    max_platform_width = 10
    for _ in range(10):
        n = random.randint(5, 10)
        heights = sorted([random.randint(1, 10) for _ in range(n)])
        widths = [random.randint(1, max_platform_width) for _ in range(n)]
        case = (n, max_platform_width, heights, widths, "Expected: TBD platforms, Total Height: TBD, Sculptures per Platform: TBD")
        cases.append(case)
    return cases


# Unimodal Valley
def generate_unimodal_valley_cases():
    cases = []
    max_platform_width = 10
    for _ in range(10):
        n = random.randint(5, 10)
        peak = random.randint(1, 10)
        left = sorted([random.randint(1, peak) for _ in range(n // 2)], reverse=True)
        right = sorted([random.randint(1, peak) for _ in range(n - len(left))])
        heights = left + right
        widths = [random.randint(1, max_platform_width) for _ in range(n)]
        case = (n, max_platform_width, heights, widths, "Expected: TBD platforms, Total Height: TBD, Sculptures per Platform: TBD")
        cases.append(case)
    return cases


# Unimodal Peak
def generate_unimodal_peak_cases():
    cases = []
    max_platform_width = 10
    for _ in range(10):
        n = random.randint(5, 10)
        peak = random.randint(1, 10)
        left = sorted([random.randint(1, peak) for _ in range(n // 2)])
        right = sorted([random.randint(1, peak) for _ in range(n - len(left))], reverse=True)
        heights = left + right
        widths = [random.randint(1, max_platform_width) for _ in range(n)]
        case = (n, max_platform_width, heights, widths, "Expected: TBD platforms, Total Height: TBD, Sculptures per Platform: TBD")
        cases.append(case)
    return cases


# Completely Random Heights
def generate_random_cases():
    cases = []
    max_platform_width = 100
    for _ in range(10):
        n = random.randint(5, 10)
        heights = [random.randint(1, 100) for _ in range(n)]
        widths = [random.randint(1, max_platform_width) for _ in range(n)]
        case = (n, max_platform_width, heights, widths, "Expected: TBD platforms, Total Height: TBD, Sculptures per Platform: TBD")
        cases.append(case)
    return cases


# Random Heights (Range 1-25)
def generate_random_heights_cases():
    cases = []
    max_platform_width = 25
    for _ in range(10):
        n = random.randint(5, 10)
        heights = [random.randint(1, 25) for _ in range(n)]
        widths = [random.randint(1, max_platform_width) for _ in range(n)]
        case = (n, max_platform_width, heights, widths, "Expected: TBD platforms, Total Height: TBD, Sculptures per Platform: TBD")
        cases.append(case)
    return cases
```

```python
# Collect all cases into one list
def collect_all_cases():
    all_cases = []
    
    # Add each scenario
    all_cases.extend(generate_non_increasing_cases())
    all_cases.extend(generate_increasing_cases())
    all_cases.extend(generate_unimodal_valley_cases())
    all_cases.extend(generate_unimodal_peak_cases())
    all_cases.extend(generate_random_cases())
    all_cases.extend(generate_random_heights_cases())
    
    return all_cases

# Generate all test cases
test_cases = collect_all_cases()

# Print the output in the requested format
print("test_cases = [")
for case in test_cases:
    print(f"    {case},")
print("]")
```

```python
import pandas as pd

# Update: Include sequence type
def collect_all_cases_with_types():
    all_cases = []
    all_cases.extend((case + ("Non-Increasing",) for case in generate_non_increasing_cases()))
    all_cases.extend((case + ("Increasing",) for case in generate_increasing_cases()))
    all_cases.extend((case + ("Unimodal Valley",) for case in generate_unimodal_valley_cases()))
    all_cases.extend((case + ("Unimodal Peak",) for case in generate_unimodal_peak_cases()))
    all_cases.extend((case + ("Random (1-100)",) for case in generate_random_cases()))
    all_cases.extend((case + ("Random (1-25)",) for case in generate_random_heights_cases()))
    return all_cases

test_cases = collect_all_cases_with_types()

def run_algorithms_and_tabulate(test_cases):
    """
    Runs Program2 (Greedy-Unimodal) and Program5B (Dynamic-Optimal) for all test cases and tabulates the results.
    """
    results = []  # To store the tabulated results

    for idx, (n, W, heights, widths, _, seq_type) in enumerate(test_cases):
        # Run Program2
        try:
            platforms_p2, height_p2, sculptures_p2 = program2(n, W, heights, widths)
        except Exception:
            platforms_p2, height_p2, sculptures_p2 = "Error", "Error", "Error"

        # Run Program5B
        try:
            platforms_p5b, height_p5b, sculptures_p5b = program5B(n, W, heights, widths)
        except Exception:
            platforms_p5b, height_p5b, sculptures_p5b = "Error", "Error", "Error"

        # Add results to the table
        results.append({
            "TestCase#": idx + 1,
            "Sequence Type": seq_type,
            "Input": f"n={n}, W={W}",
            "P2 Platforms": platforms_p2,
            "P5B Platforms": platforms_p5b,
            "P2 Height": height_p2,
            "P5B Height": height_p5b,
            "Cost Diff": "Yes" if (platforms_p2, height_p2) != (platforms_p5b, height_p5b) else "No",
        })

    # Convert results to a DataFrame
    df = pd.DataFrame(results)
    return df

# Generate and tabulate
df_results = run_algorithms_and_tabulate(test_cases)

# Ensure "Sequence Type" follows the desired order
desired_order = ["Non-Increasing", "Increasing", "Unimodal Valley", "Unimodal Peak", "Random (1-100)", "Random (1-25)"]
df_results["Sequence Type"] = pd.Categorical(df_results["Sequence Type"], categories=desired_order, ordered=True)

# Display configuration
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Group and display
# Group and display
print("Results by Sequence Type:")
for seq_type, group in df_results.groupby("Sequence Type", observed=True):  # Explicitly set observed=True
    print(f"\nSequence Type: {seq_type}")
    print(group[["TestCase#", "Input", "P2 Platforms", "P5B Platforms", "P2 Height", "P5B Height", "Cost Diff"]])


# Optional: Save results to a CSV
df_results.to_csv("algorithm_comparison_results.csv", index=False)
```
