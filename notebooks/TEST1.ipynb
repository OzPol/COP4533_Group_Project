```python
import matplotlib.pyplot as plt
import time
import random

def generate_data_constant_width(n):
    """Generate data with random heights and constant widths (all widths = 1)."""
    max_platform_width = 10
    widths = [1] * n  # All widths set to 1
    heights = [random.randint(1, 10) for _ in range(n)]  # Heights between 1 and 10
    return n, max_platform_width, heights, widths

def measure_runtime_with_fixed_data(program, data_set):
    """Measure the runtime for a given program using a fixed dataset."""
    runtimes = []
    for n, max_platform_width, heights, widths in data_set:
        start_time = time.time()
        program(n, max_platform_width, heights, widths)
        end_time = time.time()
        runtimes.append(end_time - start_time)
    return runtimes

# Generate one consistent dataset
input_sizes = range(5, 31, 5)  # Sample sizes from 10 to 50 in increments of 5
data_set = [generate_data_constant_width(n) for n in input_sizes]

# Measure runtimes for each program using the same dataset
runtimes_program3 = measure_runtime_with_fixed_data(program3, data_set)
runtimes_program4 = measure_runtime_with_fixed_data(program4, data_set)
runtimes_program5A = measure_runtime_with_fixed_data(program5A, data_set)
runtimes_program5B = measure_runtime_with_fixed_data(program5B, data_set)

# Plot individual runtimes
def plot_runtime(input_sizes, runtimes, program_name, color):
    """Plot runtime with fixed y-axis for consistency across programs."""
    plt.plot(input_sizes, runtimes, label=program_name, marker="o", color=color)
    plt.title(f"Running Time of {program_name}")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Running Time (seconds)")
    plt.ylim(0, 20)  # Fixed y-axis range to 0.05 seconds
    plt.legend()
    plt.grid(True)

plt.figure(figsize=(6, 4))
plot_runtime(input_sizes, runtimes_program3, "Program3", "purple")
plt.show()

plt.figure(figsize=(6, 4))
plot_runtime(input_sizes, runtimes_program4, "Program4", "blue")
plt.show()

plt.figure(figsize=(6, 4))
plot_runtime(input_sizes, runtimes_program5A, "Program5A", "orange")
plt.show()

plt.figure(figsize=(6, 4))
plot_runtime(input_sizes, runtimes_program5B, "Program5B", "green")
plt.show()

# Overlay plot for comparing all programs
plt.figure(figsize=(6, 4))
plt.plot(input_sizes, runtimes_program3, label="Program3", marker="o", color="purple")
plt.plot(input_sizes, runtimes_program4, label="Program4", marker="o", color="blue")
plt.plot(input_sizes, runtimes_program5A, label="Program5A", marker="o", color="orange")
plt.plot(input_sizes, runtimes_program5B, label="Program5B", marker="o", color="green")
plt.title("Running Time Comparison of Programs 3, 4, 5A, and 5B (Fixed y-axis)")
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (seconds)")
plt.ylim(0, 20)  # Fixed y-axis 
plt.legend()
plt.grid(True)
plt.show()
```
