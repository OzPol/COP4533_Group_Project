```python
import matplotlib.pyplot as plt
import time
import random

# Assume the function definitions for program3, program4, program5A, program5B are available

def generate_data(n):
    """Generate data with specified width, platform width, and height constraints."""
    max_platform_width = 100
    widths = [random.randint(1, 100) for _ in range(n)]
    heights = [random.randint(1, 1000) for _ in range(n)]
    return n, max_platform_width, heights, widths

def measure_runtime(program, input_sizes):
    """Measure the runtime for a given program across specified input sizes."""
    runtimes = []
    for n in input_sizes:
        n, max_platform_width, heights, widths = generate_data(n)
        start_time = time.time()
        program(n, max_platform_width, heights, widths)
        end_time = time.time()
        runtimes.append(end_time - start_time)
    return runtimes

# Define input sizes
input_sizes_program3 = range(1, 49)  # For Program3, limited to 48
input_sizes_others = range(100, 501, 50)  # For Programs 4, 5A, 5B
```
