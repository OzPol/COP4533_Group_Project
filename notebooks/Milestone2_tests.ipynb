```python
def run_milestone2_tests(algorithm, algorithm_name="Algorithm"):
    """
    Runs all test cases on a specified algorithm (e.g., program3, program4, program5A, or program5B).
    
    Parameters:
    algorithm: The algorithm function to test.
    algorithm_name: A string name for the algorithm for display purposes.
    """
    print(f"\nRunning tests for {algorithm_name}...\n")
    
    for i, (n, W, heights, widths, description) in enumerate(test_cases):
        print(f"Test Case {i + 1}:\n{description}")
        try:
            platforms, total_height, sculptures_per_platform = algorithm(n, W, heights, widths)
            print(f"Result:\t {platforms} platforms, Total Height: {total_height}, Sculptures per Platform: {sculptures_per_platform}\n")
        except Exception as e:
            print(f"Error: {e}\n")

# Running the tests for each program individually, one at a time.
run_milestone2_tests(program3, "Program 3")
```
